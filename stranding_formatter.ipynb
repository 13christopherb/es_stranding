```python
import pandas as pd
```

```python
strandings = pd.read_excel('/Users/christopherberglund/Projects/stranding/misc/ES 2016-2017 Admits.xlsx')
strandings2 = pd.read_excel('/Users/christopherberglund/Projects/stranding/ES 2007-2015 Admits.xlsx')
strandings2 = strandings2[['MMC_ID', 'Name', 'Age_Class', 'Strand_Date', 'Stranding_County','Stranding_Locality']]
strandings3 = pd.read_excel('/Users/christopherberglund/Projects/stranding/misc/ES_Jan-Apr_2018_Admits.xlsx')
strandings = pd.concat([strandings,strandings3])
strandings = strandings[['Animal::Name', 'Animal::TMMC_ID', 'Age_Class', 'STAR_Sightings::Strand_Date',
                         'STAR_Sightings::Locality','STAR_Sightings::County']]
strandings = strandings.rename(columns={'Animal::Name':'Name', 'Animal::TMMC_ID':'MMC_ID', 'STAR_Sightings::Strand_Date':'Strand_Date',
                                        'STAR_Sightings::Locality': 'Stranding_Locality', 'STAR_Sightings::County': 'Stranding_County'})
strandings = pd.concat([strandings, strandings2], sort=True).dropna(how='all')
strandings = strandings.sort_values('Strand_Date')
strandings.to_csv('/Volumes/Seagate/es_decade_strandings.csv', index=False)
```
